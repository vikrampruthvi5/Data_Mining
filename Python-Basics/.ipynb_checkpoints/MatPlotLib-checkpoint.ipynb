```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
```

# <font color=blue>Visualization</font>

## <font color=orange>Line Plot</font>
* Just plots points on the x-y plane

```python
# Plots line between x and y coordinates
x = [1970, 1980, 1990, 2000, 2001, 2002, 2005, 2008]
y = [3.6, 12.8, 13.6, 18.9, 19.0, 19.99, 25.34, 20.35]

plt.plot(x, y)
plt.xlabel('Year')
plt.ylabel('Population in billions')
plt.show()
```

## <font color=orange>Scatter Plot</font>
* Used to identify correlation between two attributes

```python
# use the alpha value to control the visibility of the dots
# use c to insert colors for points
# use s to control the size of the dots depending on the value
# use plt.text(values, 'Text') to plot text for the data
# use plt.grid(True) to enable grid view
z = [270.3, 908.7, 300.0, 320.98, 590.00, 120.00, 320, 323, 280.9]
plt.scatter(x, y, c = ['red', 'blue', 'green'], alpha=.8, s= z)
plt.xlabel('Year')
plt.ylabel('Population in billions')
plt.text(1999, 22, 'India')
plt.text(1980, 14.89, 'Japan')
plt.grid(True)
plt.show()
```

## <font color=orange>Histogram</font>
* Defines frequency distribution of data 
* Divided x axis into bins and find the frequency of data points in that bins

```python
x = np.array([1, 3, 2, 2, 4, 3, 4,6, 5, 4, 3, 5, 7, 9, 5, 7, 0, 9, 0, 9, 2, 3, 5, 2, 7, 5, 7, 6])
plt.figure(figsize=[5, 5])
plt.hist(x, bins = 'auto')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## <font color=red>When to select which graph?</font>
* You're a professor teaching Data Science with Python, and you want to visually assess if the grades on your exam follow a particular distribution. Which plot do you use? <font color=blue>Histogram</font>
* You're a professor in Data Analytics with Python, and you want to visually assess if longer answers on exam questions lead to higher grades. Which plot do you use? <font color=blue>Scatter plot</font>

## <font color=blue>Data Structures</font>
### <font color=orange>Dictionaries</font>
* Key Value pair
* Inside curly brackets {'Key' : 'Value'}

```python
# Hastle of creating multiple lists can be avoided using dictionaries

# Create 
# Implicit creation
capitals = {'spain': 'madrid', 'france': 'paris', 'germany': 'berlin', 'norway':'oslo' }
# Explicit creation
x = dict()
x['India'] = 'New Delhi'

# Access dict values
print(capitals['germany'])
print(x['India'])

# Print only keys in a dict
print(capitals.keys())

# Check whether the key exists : bool
print('germany' in capitals)
```

```python
# Problem with Lists
countries = ['spain', 'france', 'germany', 'norway']
capitals = ['madrid', 'paris', 'berlin', 'oslo']

# Accessing capital of germany
ind = countries.index('germany')
print(capitals[ind])
```
