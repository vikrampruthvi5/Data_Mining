### Age and Cabin attributes have missing values

```python
# Function that calculates Missing values counts of a variable
## Input : Attribute name and dataset name
def missingCounter(ds, attr):
    total_count = 0
    missing_count = 0
    data = ds[attr]
    for i in data:
        total_count += 1
        if str(i)=='nan':
            missing_count += 1
    print(attr + "\n" + "Total count \t: " + str(total_count) + "\nMissing count \t: " + str(missing_count))
```

```python
# Handling Age data
## Missing value count
missingCounter(dataset, 'Age')
```

```python
# Dropping Cabins column as it has many missing values...
dataset = dataset.drop(columns=['Cabin'], axis=1)

# Handling Age data
dataset = dataset.dropna()
```
