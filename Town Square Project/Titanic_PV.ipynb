```python
# Method to get missing data counts of each attribute

def findMissingDataCounts():
    # length of data array
    missing_values = dict()
    attrs = len(dtrain)
    cols = dtrain.columns
    for i in cols:
        non_missing_vals = dtrain[i].value_counts().sum()
        missing_values_count = attrs-non_missing_vals
        missing_values[i] = missing_values_count
    print(missing_values)
    
```

## Lets write a function for numeric data to plot distribution and count graphs
* Count of variables
* missing values
* distribution

```python
# Trying to fill age with mean values
dtrain.Age.fillna(dtrain.Age.mean(), inplace = True)
```

```python
def plot_count_dist_missing(data, bin_df, label_column, target_var, figr_size):
    fig = figr_size
    target_var = target_var
    data = data
    
#     if df_bin:
#         data = df_bin
#     else:
#         data = data

    # survived and age - count
    print("Missing data count : {}".format(dtrain[target_var].isnull().sum()))

    # survived and age - dist
    plt.subplot(3,2,1)
    sns.distplot(dtrain.loc[data[label_column] == 1][target_var], kde_kws={"label": label_column});

    # survived and age - count plot
    plt.subplot(3, 2, 2)
    sns.countplot(y=dtrain[target_var], data = data)
```

```python
fig = plt.figure(figsize=(20,30))
plot_count_dist_missing(dtrain, 0,'Survived','Age', fig)
```
